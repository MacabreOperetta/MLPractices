```python
from sklearn.cluster import AgglomerativeClustering
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset = pd.read_csv("https://raw.githubusercontent.com/jupyter/docker-demo-images/master/datasets/cluster/xclara.csv", index_col=0)
dataset.head()
```

```python
plt.scatter(dataset.V1, dataset.V2, edgecolors='face')
```

```python
model = AgglomerativeClustering(n_clusters=3, linkage='ward', affinity='euclidean')
model.fit(dataset)
```

```python
plt.scatter(dataset.V1, dataset.V2, c=model.labels_, cmap='rainbow', edgecolors='face')
```
